### Profils type

```python
import pandas as pd
from datetime import datetime
```

```python
df_profils_1=pd.read_csv('validations-sur-le-reseau-ferre-profils-horaires-par-jour-type-1er-sem.csv',delimiter=';')
df_profils_2=pd.read_csv('validations-sur-le-reseau-ferre-profils-horaires-par-jour-type-2e-sem.csv',delimiter=';')

print(df_profils_1.shape)
df_profils_1=df_profils_1.dropna()
df_profils_1=df_profils_1.loc[df_profils_1['CAT_JOUR']=="JOHV",:]
df_profils_1['ID_REFA_LDA']=df_profils_1.ID_REFA_LDA.astype(float)
df_profils_1=df_profils_1.reset_index(drop=True)
df_profils_1=df_profils_1.drop(columns=['CODE_STIF_TRNS','CODE_STIF_RES','CODE_STIF_ARRET','CAT_JOUR'])
df_profils_1=df_profils_1[['ID_REFA_LDA','LIBELLE_ARRET','TRNC_HORR_60','pourc_validations']]
print(df_profils_1.shape)

print(df_profils_2.shape)
df_profils_2=df_profils_2.dropna()
df_profils_2=df_profils_2.loc[df_profils_2['CAT_JOUR']=="JOHV",:]
df_profils_2=df_profils_2.loc[df_profils_2['ID_REFA_LDA']!="?",:]
df_profils_2['ID_REFA_LDA']=df_profils_2.ID_REFA_LDA.astype(float)
df_profils_2=df_profils_2.reset_index(drop=True)
df_profils_2=df_profils_2.drop(columns=['CODE_STIF_TRNS','CODE_STIF_RES','CODE_STIF_ARRET','CAT_JOUR'])
df_profils_2=df_profils_2[['ID_REFA_LDA','LIBELLE_ARRET','TRNC_HORR_60','pourc_validations']]
print(df_profils_2.shape)
```

```python
counterror=0
df_profils=df_profils_1
for i in range(df_profils_1.shape[0]):
    ID_REFA_LDA=df_profils_1.iloc[i,0]
    TRNC_HORR_60=df_profils_1.iloc[i,2]
    pourc_validations_1=df_profils_1.iloc[i,3]
    try:
        pourc_validations_2=df_profils_2.loc[(df_profils_2['ID_REFA_LDA']==ID_REFA_LDA) & (df_profils_2['TRNC_HORR_60']==TRNC_HORR_60),:].iloc[0,3]
    except Exception as e:
        counterror+=1
        print(e,counterror)
        pourc_validations_2=pourc_validations_1
    pourc_validations=(pourc_validations_1+pourc_validations_2)/2
    df_profils.iloc[i,3]=pourc_validations
```

### Validations par jour

```python
df_nombre_1=pd.read_csv('validations-sur-le-reseau-ferre-nombre-de-validations-par-jour-1er-sem.csv',delimiter=';')
df_nombre_2=pd.read_csv('validations-sur-le-reseau-ferre-nombre-de-validations-par-jour-2e-sem.csv',delimiter=';')
df_nombre=pd.concat([df_nombre_1,df_nombre_2])
```

```python
print(df_nombre.shape)
df_nombre=df_nombre.dropna()
print(df_nombre.shape)
df_nombre=df_nombre.loc[df_nombre['CATEGORIE_TITRE']!="NON DEFINI",:]
df_nombre['JOUR']=df_nombre.JOUR.astype(str)
df_nombre=df_nombre.reset_index(drop=True)
df_nombre=df_nombre.drop(columns=['CODE_STIF_TRNS','CODE_STIF_RES','CODE_STIF_ARRET','CATEGORIE_TITRE'])
df_nombre=df_nombre[['JOUR','ID_REFA_LDA','LIBELLE_ARRET','NB_VALD']]
```

```python
df_nombre.head()
```

```python
def week_number(row):
    d=datetime.strptime(row.JOUR, "%Y-%m-%d")
    return d.isocalendar()[1]
def week_day(row):
    d=datetime.strptime(row.JOUR, "%Y-%m-%d")
    return d.isocalendar()[2]
```

```python
df_nombre['WEEK_NUMBER']= df_nombre.apply (lambda row: week_number (row),axis=1)
df_nombre['WEEK_DAY']= df_nombre.apply (lambda row: week_day (row),axis=1)
```

```python
JOHV=[1,2,3,4,5]
Vacances=[6,7,14,15,28,29,30,31,32,33,34,35,43,44,52]
Jours_feries=['2017-01-01','2017-04-16','2017-04-17','2017-05-01','2017-05-08','2017-05-25','2017-06-05','2017-07-14','2017-08-15','2017-11-01','2017-11-11','2017-12-25']
df_nombre=df_nombre.loc[df_nombre['WEEK_DAY'].isin(JOHV),:]
df_nombre=df_nombre.loc[~df_nombre['WEEK_NUMBER'].isin(Vacances),:]
```
